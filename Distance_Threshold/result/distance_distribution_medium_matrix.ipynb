```python
# All required libraries
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns

# Settings such as to make sure the plot show in the notebook 
%matplotlib inline
sns.set(color_codes=True)
np.random.seed(sum(map(ord, "distributions")))
```

Note that distances ordered background, from smallest to largest. Largest means the paper is closer to the target (i.e. distance=1 means the target itself) 

```python
# Read first line of our small matrix file 
array_150 = []
array_150_closest = []
array_150_all = []
array_150_closest_all = []
with open("/projects/slin_prj/threshold_experiment/big_matrix/_150/150.selected_distances", 'r') as f:
    for line in f:
        distances = line.strip().split()
        array_150.append([float(d) for d in distances[1:]])
        for d in distances[1:]: 
            array_150_all.append(float(d))
with open("/projects/slin_prj/threshold_experiment/big_matrix/_150/150.closest_papers", 'r') as f:
    for line in f:
        distances = line.strip().split()
        array_150_closest.append([float(d) for d in distances[1:]])
        for d in distances[1:]: 
            array_150_closest_all.append(float(d))
```

```python
plot = sns.distplot(array_150[10], kde=False)
plot2 = sns.distplot(array_150_closest[10], kde=False)
axes = plot.axes
axes.set_xlim(-1, 1)
```

```python
# Zoom into the plot
plot = sns.distplot(array_150[10], kde=False, bins=40)
plot2 = sns.distplot(array_150_closest[10], kde=False, bins=40)
axes = plot.axes
axes.set_xlim(0.5, 1)
axes.set_ylim(0, 100)
```
